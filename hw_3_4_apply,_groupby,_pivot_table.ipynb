1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
branch_profit = df.groupby('Branch')['gross income'].sum()

most_profitable_branch = branch_profit.idxmax()
highest_profit = branch_profit.max()

print(most_profitable_branch)
print(highest_profit)
```

2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  
Знайдіть відповідь за допомогою функціоналу Pandas.

```python
branch_data = df[df['Branch'] == most_profitable_branch]
most_profitable_city = branch_data['City'].unique()[0]

most_profitable_city
```

3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index = 'Branch', columns = 'Product line', values = 'Invoice ID', aggfunc='count', fill_value=0)
invoices_by_category
```

3.2. Викристовуючи змінну `invoices_by_category` дайте відповідь програмно (тобто значення треба не просто знайти очима, а вивести за допомогою коду), в якому філіалі магазину (`Branch`) найбільше інвойсів із покупкою товарів категорії "Електронні аксесуари" (`Electronic accessories`)?


```python
most_electronics = invoices_by_category['Electronic accessories'].idxmax()
max_invoices = invoices_by_category['Electronic accessories'].max()

print(most_electronics)
print(max_invoices)
```

# **Відповіді**
1. Якщо певні дні є більш популярними для покупок, компанія може ввести спеціальні акції або додаткові знижки в інші, менш активні дні, щоб збільшити потік клієнтів і рівномірно розподілити навантаження на магазини.

3. Виявлення товарних категорій або філіалів із низькими рейтингами дозволить бізнесу покращити обслуговування, асортимент або якість продукції.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Day of Week'] = df['Date'].dt.day_name()

purchases_by_day = df.groupby(['Branch', 'Day of Week'])['Invoice ID'].count()

purchases_by_day
```

```python
avg_total_by_payment = df.groupby('Payment')['Total'].mean()

avg_total_by_payment
```

```python
avg_rating_by_category_branch = df.groupby(['Branch', 'Product line'])['Rating'].mean()

avg_rating_by_category_branch
```
